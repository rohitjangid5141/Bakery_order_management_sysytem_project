```python
import pandas as pd
from sqlalchemy import create_engine

class BakeryOrder:
    order_id_count = 101
    def __init__(self):
        self.orders = pd.DataFrame(columns=["Order_ID","Customer_name",
                                    "Order_qty", "Item"]).reset_index(drop=True)
        self.Menu()
    def Menu(self):
        print("""
        1. Enter 1 For Add Order.
        2. Enter 2 For View Order.
        3. Enter 3 ForUpdate Order.
        4. Enter Anything For Exit.""")
        
        user_in = input("Enter The No. ")
        
        if user_in == '1':
            self.Add_order()
        elif user_in=='2':
            self.View_order()
        elif user_in=='3':
            self.update_order()
        elif user_in=='4':
            self.Save_to_Database()
        else:
            exit
            
    def Add_order(self):
        customer_name = input("Enter Your Name. ")
        order_qty = int(input("Enter Your Order Quantity. "))
        order_name = input("Enter Your Order Name. ")
        order_id = BakeryOrder.order_id_count
        BakeryOrder.order_id_count += 1
        order_data = iter([{"Order_ID" : order_id, "Customer_name":customer_name,
                     "Order_qty":order_qty,"Item":order_name}])
        for item in order_data:
            self.orders=self.orders.append(item,ignore_index=True)
        self.__Save_to_Database()
        print("Order Placed Successfully")
        self.Menu()
          
    def View_order(self):
        if self.orders.empty:
            print("No Orders Are Added Please Add Orders.")
            self.Menu()
        else:
            print(self.orders)
            self.Menu()
    
    def __Save_to_Database(self):
        username = 'root'
        password = '123'
        host = 'localhost'
        database_name = 'bakery'
        
        connection_str = f"mysql+pymysql://root:123@localhost/bakery"
        engine = create_engine(connection_str)
        self.orders.to_sql('order_data', engine, index=False, if_exists='append')
        self.Menu()
        
    def update_order(self):
        choice=input("""
        1. Enter 1 For Updating Order Quantity.
        2. Enter 2 For Updating Order Name.
        3. Enter Anything To Exit.""")
        
        if choice == '1':
            print("Order IDs",self.orders["Order_ID"])
            order_id = int(input("Enter Your Order Id. "))
            if order_id in self.orders['Order_ID'].values and choice=='1':
                quantity = input("Add Quantity. ")
                self.orders.loc[self.orders["Order_ID"]==order_id,'Order_qty'] = quantity
                print("Order Updated SuccessFully. ")
                self.Menu()
            else:
                print("Order Not Found. ")
        elif choice=='2':
            print("Order IDs",self.orders["Order_ID"])
            order_id = int(input("Enter Your Order Id. "))
            if order_id in self.orders['Order_ID'].values and choice=='2':
                item = input("Update Item. ")
                self.orders.loc[self.orders["Order_ID"]==order_id,'Item'] = item
                print("Order Updated SuccessFully. ")
                self.Menu()
        else:
            exit
            self.Menu()
            
            
```
